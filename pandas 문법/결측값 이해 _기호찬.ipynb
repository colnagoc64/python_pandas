## 결측값 NaN

결측값이 많아지면 데이터값이 외곡되고 편향적으로 바뀌기 떄문에 제거해야될 1순위

결측값을 처리하는 방법 : 전처리과정 :prepocessing ,prunning
                         정규화과정



```python
import pandas as pd
```

```python
idol=pd.read_csv('idol.csv')
idol
```

```python
idol.info() 
# non-null 정상값 정상치 이다. 
# 15 non-null 이 기준이되어시 13 ,14 non-null은 이상이 있을 수있다 
```

```python
idol.isna() 
# 빠진 값이 있냐? 했을때
# true 값이면 빠진값이다 false 빠진값이 아니다.
# 따라서 true 값이면 결측지값으로 판단된다 한번 확인해야한다.
```

```python
idol[['그룹']].isnull()
# 빠진 값이 있냐? 했을때
# true 값이면 빠진값이다 false 빠진값이 아니다.
# 따라서 true 값이면 결측지값으로 판단된다 한번 확인해야한다.
# 위에 해석한 흐름과 같음
```

```python
idol['그룹'][idol['그룹'].isnull()]
# NaN값만 찾아내기
```

```python
idol['그룹'][idol['그룹'].isna()]
# NaN값만 찾아내기
```

```python
idol[['그룹']].notnull()
```

```python
idol['그룹'][idol['그룹'].notnull()]
# NaN 아니값만 출력
```

```python
idol.loc[idol['그룹'].notnull() ,['키','혈액형']]
```

# merge 

```python
cust=pd.DataFrame({
    '고객번호':[1001,1002,1003,1004,1005,1006,1007],
    '이름':['가','나','다','라','마','바','사']
}, columns=['고객번호','이름'])
cust
```

```python
order=pd.DataFrame({
    '고객번호':[1001,1001,1005,1004,1008,1006,1007],
    '금액':[10000,20000,300000,40000,5000,6000,7000]
}, columns=['고객번호','금액'])
order
```

```python
pd.merge(cust,order)
```

```python
pd.merge(cust,order,how='outer') # outer 기본값
```

```python
pd.merge(cust,order,how='left')
```

```python
pd.merge(cust,order,how='right')
```

```python
df1= pd.DataFrame({
    '품종':['setosa','setosa','virginica','virginica'],
    '꽃잎길이':[1.4,1.3,1.5,1.3]},
    columns=['품종','꽃잎길이'])
df1
```

```python
df2= pd.DataFrame({
    '품종':['setosa','setosa','virginica','virginica'],
    '꽃잎길이':[0.4,0.3,0.5,0.3]},
    columns=['품종','꽃잎길이'])
df2
```

```python
pd.merge(df1,df2, how='outer')
```

```python
df3= pd.DataFrame({
    '고객명':['춘향','춘향','몽룡'],
    '날짜':['2019-01-01','2019-01-02','2019-01-03'],
    '데이터':['20000','30000','100000']})
df3
```

```python
df4= pd.DataFrame({
    '고객명':['춘향','몽룡'],
    '데이터':['여자','남자']})
df4
```

```python
pd.merge(df3,df4, on ='고객명')
```
