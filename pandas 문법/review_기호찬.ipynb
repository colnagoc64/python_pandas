```python
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    
    return year +'-'+ month + '-' + day
```

```python
df1= pd.DataFrame({
    '이름':['A','B','D'],
    '성별':['남자','여자','남자']})
df1
```

```python
df2= pd.DataFrame({
    '이름':['A','B','C'],
    '키':[180,160,183]})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how='left') # df1의 key 값을 기준으로 하기대문에 df2 에 'D'에 키정보가없어서 nan 
```

```python
pd.merge(df1,df2, how='right')
```

```python
pd.merge(df1,df2,how='outer')
```
