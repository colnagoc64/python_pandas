```python
#라벨 그래프옆에 주석

x=['Mon','Tue','Wed','Thur','Fri','sat','sun']
y=[13,16,15,18,16,17,16]
plt.plot(x,y)
plt.xlabel('Daily')
plt.ylabel('No. of Books sold')
```

```python
#그래프 1개 더 추가
x=['Mon','Tue','Wed','Thur','Fri','sat','sun']
y1=[13,16,15,18,16,17,16]
y2=[17,14,17,16,15,15,14]
plt.plot(x,y1) #파란색
plt.plot(x,y2) #노란색
plt.xlabel('Daily')
plt.ylabel('No. of Books sold')
```

```python
x=['Mon','Tue','Wed','Thur','Fri','sat','sun']
y1=[13,16,15,18,16,17,16]
y2=[17,14,17,16,15,15,14]
plt.plot(x,y1,label='sold') #파란색  label을 붙여주고
plt.plot(x,y2,label='on shelves') #노란색
plt.xlabel('Daily')
plt.ylabel('No. of Books sold')
plt.legend(loc="upper left") # 이것도 붙여줘야 위에 주석까지뜬다
```

```python
x=['Mon','Tue','Wed','Thur','Fri','sat','sun']
y1=[13,16,15,18,16,17,16]
y2=[17,14,17,16,15,15,14]
plt.plot(x,y1,label='sold') 
plt.plot(x,y2,label='on shelves') 
plt.xlabel('Daily')
plt.ylabel('No. of Books sold')
plt.legend(loc="upper left") 
plt.title('Le petit Prince : sold & Left')
```

```python
x=[1,2,3,4,5,]
y=[1,4,9,16,25]
plt.plot(x,y,'bo') #bo => o 모양 점 표시
plt.show()
```

```python
x=[1,2,3,4,5,]
y=[1,4,9,16,25]
plt.plot(x,y,'bo')
plt.axis([0,6,0,30]) # x축과 y축 범위 지정  (앞에 2개 x축범위 뒤에 2개 y축 범위)
plt.show()
```

```python
import numpy as np
points = np.array([[1,1],[1,2],[1,3],[2,1],[2,2],[2,3],[3,1],[3,2],[3,3]])
p= np.array([2.5,2])
plt.plot(points[:,0],points[:,1],"ro") # 행은 배열을 슬라이스 하여 모든 행( :)을 취하고 두 번째 열( 1)을 유지합니다.
plt.plot(p[0],p[1],"bo")
```

```python
x = np.linspace(0,5,10)
y = x**2
plt.plot(x,y);
```

```python
x = np.linspace(0,5,20)
y = x**2
plt.plot(x,y,"bo-",linewidth=3,markersize=5); #bo'-' 를 넣는 이유는 선을 표현하기위해
```

```python
plt.plot(x,y,"gs-",linewidth=1,markersize=7);
```

```python
x= np.linspace(0,10,20)
y1= x**2.0
y2= x*1.5
plt.plot(x,y1,"bo-",linewidth=2,markersize=5,label="First")
plt.plot(x,y2,"gs-",linewidth=2,markersize=5,label="Second")
plt.xlabel("x")
plt.xlabel("y")
plt.axis([-0.5,10.5,-5,105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```

```python
x= np.logspace(-1,1,20)
y1= x**2.0
y2= x**1.5
plt.plot(x,y1,"bo-",linewidth=2,markersize=5,label="First")
plt.plot(x,y2,"gs-",linewidth=2,markersize=5,label="Second")
plt.xlabel("x")
plt.xlabel("y")
plt.axis([-0.5,10.5,-5,105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```
